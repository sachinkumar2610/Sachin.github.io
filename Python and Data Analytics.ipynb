```python
# matplotlib.pyplot function is used for data visualisation
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([0, 6])
y = np.array([0, 250])
```

```python
x, y
```

```python
# plot function is used to make a line graph
plt.plot(x,y)
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# 'o' signifies the dot where the numbers occur
plt.plot(x,y,'o')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
x = np.array([1,2,6,8])
y = np.array([3,8,1,10])
```

```python
plt.plot(x,y)
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# '*' is used to put this special character instead of dot and 'marker' function is used to put line and dots together on the graph 
plt.plot(x,y,marker='*')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# 'D' is used to replace the points with diamonds
plt.plot(x,y,marker='D')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# 'H' is used to replace the points with hexagon
plt.plot(x,y,marker='H')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# ':' function is used to replace the plane line into dotted line
# 'r' function is a red color which changes the line into red
plt.plot(x,y,'o:r')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# '--' function is used to replace the plane line into dashed line
plt.plot(x,y,'o--r')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# '-.' function is used to replace the plane line into dotted and dashed line
plt.plot(x,y,'o-.r')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# 'ms' function is marker size which is used to increase the size of dot
plt.plot(x,y,'o--', ms=20)
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# 'mec' function is marker edge color which covers the outline of dot into specified color
plt.plot(x,y,'o--',ms=20, mec='r')
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```

```python
# 'linestyle' or 'ls' function is used to make the plane line into any specified line you want. 
# 'linewidth' function is used to increase width of the line
plt.plot(x,y, ls='--', color='r', linewidth=20)
plt.xlabel('Product feature')
plt.ylabel('Price')
plt.title('Product feature V/s Price')
```
